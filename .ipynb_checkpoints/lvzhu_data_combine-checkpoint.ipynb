```python
import os
import glob
import numpy as np
import pandas as pd
import anndata as ad
from fcsparser import parse
```

```python
input_base = "/public/users/xueyupeng/VZV/lvzhu/subtype"
output_base = "/public/users/xueyupeng/VZV/lvzhu/subtype_combined"

os.makedirs(output_base, exist_ok=True)
```

```python
def read_fcs(file):
    meta, data = parse(file, reformat_meta=True)
    return data
```

```python
def process_one_folder(folder_path):
    files = glob.glob(os.path.join(folder_path, "*.fcs"))
    
    if len(files) == 0:
        return None
    
    dfs = []
    
    for f in files:
        df = read_fcs(f)
        
        # 👉 添加样本标记
        sample_name = os.path.basename(f).replace(".fcs", "")
        df["sample_id"] = sample_name
        
        dfs.append(df)
    
    df_all = pd.concat(dfs, axis=0)
    
    # 👉 构建 AnnData
    X = df_all.drop(columns=["sample_id"]).values
    var = pd.DataFrame(index=df_all.drop(columns=["sample_id"]).columns)
    obs = pd.DataFrame(index=[f"cell_{i}" for i in range(df_all.shape[0])])    
    obs["sample_id"] = df_all["sample_id"].values
    
    adata = ad.AnnData(X=X, obs=obs, var=var)
    
    return adata
```

```python
# ===== 主循环 =====
for subdir in os.listdir(input_base):
    folder_path = os.path.join(input_base, subdir)
    if not os.path.isdir(folder_path):
        continue    
    adata = process_one_folder(folder_path)
    if adata is None:
        continue
    out_file = os.path.join(output_base, f"{subdir}.h5ad")
    adata.write(out_file)
    print("Saved:", out_file)
```
