```python
cluster_anno = {
    "0": "NK",
    "1": "CD8T",
    "2": "CD8T",
    "3": "CD4T",
    "7": "CD4T",
    "8": "CD4T",
    "9": "CD4T",
    "10": "B",
    "5": "Plasma",
    "6": "DC",
    "4": "Mono"
}
```

```python
adata_all.obs["celltype"] = adata_all.obs["leiden"].map(cluster_anno)
```

```python
sc.pl.umap(
    adata_all,
    color="celltype",
    legend_loc="on data",
    frameon=False,
    size=5,
    title="UMAP - Celltype"
)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 假设 marker 列表
markers = ['pSTAT3_176Yb', 'pSTAT1_175Lu']  # 你实际用的 marker 列表
timepoints = sorted(adata_all.obs['Timepoint'].unique())

# 1️⃣ 先取 V1Day0 的数据，计算每个 marker 中位数和 SD
v1day0 = adata_all[adata_all.obs['Timepoint'] == 'V1Day0']

marker_stats = {}
for marker in markers:
    median = v1day0.obs[marker].median()
    sd = v1day0.obs[marker].std()
    lower = median - 2*sd
    upper = median + 2*sd
    # 定义阴性 ≤ 中位数+2SD，阳性 > 中位数+2SD
    marker_stats[marker] = {'median': median, 'sd': sd, 'upper_threshold': median + 2*sd}

# 2️⃣ 对所有数据，给每个 marker 打标签: 'Positive' / 'Negative'
for marker in markers:
    threshold = marker_stats[marker]['upper_threshold']
    adata_all.obs[f'{marker}_status'] = np.where(
        adata_all.obs[marker] > threshold,
        'Positive',
        'Negative'
    )

# 3️⃣ 计算每个 Sample + Timepoint + Marker 的阳性细胞比例和阳性表达中位数
records = []
for sample in adata_all.obs['Sample'].unique():
    for tp in timepoints:
        subset = adata_all.obs[(adata_all.obs['Sample'] == sample) & (adata_all.obs['Timepoint'] == tp)]
        for marker in markers:
            total_cells = subset.shape[0]
            positive_cells = subset[subset[f'{marker}_status']=='Positive']
            pos_count = positive_cells.shape[0]
            pos_ratio = pos_count / total_cells * 100 if total_cells > 0 else np.nan
            pos_median_expr = positive_cells[marker].median() if pos_count > 0 else np.nan
            
            records.append({
                'Sample': sample,
                'Timepoint': tp,
                'Marker': marker,
                'Total_Cells': total_cells,
                'Positive_Cells': pos_count,
                'Positive_Ratio_%': pos_ratio,
                'Positive_Median_Expr': pos_median_expr
            })

marker_summary = pd.DataFrame(records)

# 保存表格
marker_summary.to_csv("/public/users/xueyupeng/VZV/lvzhu/record/marker_positive_summary.csv", index=False)
print("✓ Marker阳性细胞比例和表达中位数已保存")

# 4️⃣ 可视化趋势（小提琴+散点图）
for marker in markers:
    plt.figure(figsize=(7,5))
    plot_df = marker_summary[marker_summary['Marker'] == marker]
    
    sns.violinplot(
        data=plot_df,
        x='Timepoint',
        y='Positive_Ratio_%',
        inner=None,
        color='lightgray'
    )
    
    sns.stripplot(
        data=plot_df,
        x='Timepoint',
        y='Positive_Ratio_%',
        hue='Sample',
        dodge=False,
        size=5,
        alpha=0.8
    )
    
    plt.title(f"{marker} Positive Ratio Trend")
    plt.ylabel("Positive Ratio (%)")
    plt.xlabel("Timepoint")
    plt.legend([],[], frameon=False)  # 不显示图例
    plt.tight_layout()
    plt.show()
    
    # 阳性表达中位数
    plt.figure(figsize=(7,5))
    sns.violinplot(
        data=plot_df,
        x='Timepoint',
        y='Positive_Median_Expr',
        inner=None,
        color='lightgray'
    )
    sns.stripplot(
        data=plot_df,
        x='Timepoint',
        y='Positive_Median_Expr',
        hue='Sample',
        dodge=False,
        size=5,
        alpha=0.8
    )
    plt.title(f"{marker} Positive Median Expression Trend")
    plt.ylabel("Median Expression (Positive Cells)")
    plt.xlabel("Timepoint")
    plt.legend([],[], frameon=False)
    plt.tight_layout()
    plt.show()
```
